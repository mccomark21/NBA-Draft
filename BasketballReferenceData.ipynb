# Acquire List of Players Who Were Drafted

```python
# Import Dependencies
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# Create df which will be used to catch all player data

DraftedPlayers = pd.DataFrame(columns=['Pk','Tm','Player','College','Year'])
```

```python
# Create list of Years to scrape
Year = [2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]

# For loop to cycle through all the years specified in the list
## Will only return the columns Pick, Team Drafted, Player Name, College, and Year

### Link 1: http://savvastjortjoglou.com/nba-draft-part01-scraping.html
### Link 2: 

for i in range(len(Year)):
    url = 'http://www.basketball-reference.com/draft/NBA_{}.html'.format(Year[i])
   
    html = urlopen(url)
    soup = BeautifulSoup(html)
    
    column_headers = [th.getText() for th in 
                  soup.findAll('tr', limit=2)[1].findAll('th')]
    
    column_headers.remove("Rk")
    
    data_rows = soup.findAll('tr')[2:]  # skip the first 2 header rows
    
    player_data = [[td.getText() for td in data_rows[i].findAll('td')]
                for i in range(len(data_rows))]
    
    df = pd.DataFrame(player_data, columns=column_headers)
    df = df[['Pk','Tm','Player','College']]

    df = df.replace({'College': {'': np.nan}}).dropna(subset=['College'])
    df['Year'] = Year[i]
    DraftedPlayers = pd.concat([DraftedPlayers, df])
```

```python
# Reset the index for the DF
DraftedPlayers = DraftedPlayers.reset_index(drop=True)

# Set the path
path = "D:\\Fall Cluster Course\\Data\\"

# Write to CSV
DraftedPlayers.to_csv(path+'DraftedPlayers.csv')

DraftedPlayers
```

```python
##### PER GAME TABLE #####

# Isolate Per Game Table

Table = soup.findAll('table',{'id':'players_per_game'})

# Create List of Data-stats from HTML table

DataStats = []

for itm in Table[0].findAll("th"):
    DataStats.append(itm['data-stat'])

unwanted = {'season','sos-dum','awards-dum'}
DataStats = [e for e in DataStats if e not in unwanted]        
# DataStats

# Create dictionary with column headers as keys and stats as values

dicts = {}

# Create Dictionary Key and Values for Season
season = []
season_bs4 = (Table[0].findAll("th", {"scope":"row"},{"data-stat":"season"} ))
for itm in season_bs4:
        season.append(itm.getText())
dicts['season'] = season

# Create Dictionary Key and Values for remaining stats

for i in range(0,len(DataStats)):
   for elem in DataStats:
    temp = []
    temp_bs4 = (Table[0].findAll("td", {"data-stat":[elem]}))
    for itm in temp_bs4:
        temp.append(itm.getText())
    dicts[elem] = temp
dicts
```

```python
# Convert dictionary to data frame

sampDF = pd.DataFrame(dicts)
sampDF
```
